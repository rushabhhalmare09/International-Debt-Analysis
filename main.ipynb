#### *Joining two Dataset*

```python

# fileSource.merge(fileDestination, left_on='lkey', right_on='rkey')  -- merging data into source

merge_file = file.merge(file1, left_on='Series Code', right_on='Code' )
```

#### **Creating a new Column and adding all the debt yearwise into single column Debt**

```python
# df['Sum']=df.iloc[:,1:3].sum(axis=1) print(df)

merge_file['debt']=merge_file.iloc[:,6:64].sum(axis=1)
```
